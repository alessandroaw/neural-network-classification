```python
from PIL import Image
import numpy as np
import pandas as pd
import random
```

```python
#try load training set
train_filename="el4233-2018-2019-02-ann-klasifikasi-train.csv"
test_filename="el4233-2018-2019-02-ann-klasifikasi-test.csv"
#training=np.genfromtxt(train_filename,delimiter=',')
```

```python
training = pd.read_csv(train_filename,header=None)
training.head()
```

```python
training_X = training.iloc[:,1:3].values
training_z = training.iloc[:,3].values
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10)) 
plt.scatter(training[1], training[2], c=training_z)
plt.xlabel("x")
plt.ylabel("y")
```

```python
from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(train[0],train[1])
X_train,X_val,z_train,z_val=train_test_split(training_X,training_z,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_scale = sc.fit_transform(X_train)
X_val_scale = sc.fit_transform(X_val)
# Assign the scaled data to a DataFrame & use the index and columns arguments to keep your original indices and column names:
#X_train = pd.DataFrame(X_train_array, index=X_train.index, columns=X_train.columns)
# Center test data with the  &  computed (fitted) on training data

#X_test_array = sc.transform(X_test.values)
#X_test = pd.DataFrame(X_test_array, index=X_test.index, columns=X_test.column
```

```python
from sklearn.neural_network import MLPClassifier
clf=MLPClassifier(alpha=1e-5, solver='adam',
                  hidden_layer_sizes=(10, 50, 100), random_state=1,
                  verbose=True, learning_rate='adaptive')
clf.fit(X_train_scale, z_train)
```

```python
predict_train = clf.predict(X_val)
print(clf.score(X_val_scale, z_val))
print (clf.n_layers_)
print (clf.n_iter_)
print (clf.loss_)
```
